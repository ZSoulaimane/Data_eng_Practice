```python
columns_to_drop = ['Zone_complementaire', 'anciennete_locataire_homogene', 'nombre_logements']
columns_to_drop
```

```python
copy_df = copy_df.drop(columns=columns_to_drop)
copy_df
```

```python
import re
```

```python
def extract_type_habitat(row):
  if pd.isnull(row['Type_habitat']) and not pd.isnull(row['nombre_pieces_homogene']):
    var = re.search(r'([A-Za-z]*\s)', str(row['nombre_pieces_homogene'])).group(1).replace(" ","")
    return 'Appartement' if var == 'Appart' else var
  else :
    return row['Type_habitat']
```

```python
copy_df['Type_habitat'] = copy_df.apply(lambda row: extract_type_habitat(row), axis=1)
```

```python
copy_df
```

```python
copy_df['Type_habitat'].unique()
```

```python
def extract_epoque(value):
  if pd.isnull(value):
    return value
  else :
    return " ".join(value.split(" ")[1:])
```

```python
copy_df['epoque_construction_homogene'] = copy_df['epoque_construction_homogene'].apply(lambda value: extract_epoque(value))
```

```python
copy_df
```

```python
def change_method(value):

  if pd.isnull(value):
    return value

  else :
    return re.search(r'([A-Za-z\s]*)', value).group()
```

```python
copy_df['methodologie_production'] = copy_df['methodologie_production'].apply(lambda value : change_method(value))
```

```python
copy_df
```

```python
copy_df['loyer_1_decile'] = copy_df['loyer_1_decile'].str.replace(',','.')
copy_df['loyer_1_quartile'] = copy_df['loyer_1_quartile'].str.replace(',','.')
copy_df['loyer_median'] = copy_df['loyer_median'].str.replace(',','.')
copy_df['loyer_3_quartile'] = copy_df['loyer_3_quartile'].str.replace(',','.')
copy_df['loyer_9_decile'] = copy_df['loyer_9_decile'].str.replace(',','.')
copy_df['loyer_moyen'] = copy_df['loyer_moyen'].str.replace(',','.')
```

```python
copy_df['loyer_1_decile'] = copy_df['loyer_1_decile'].astype(float, errors = 'raise')
copy_df['loyer_1_quartile'] = copy_df['loyer_1_quartile'].astype(float, errors = 'raise')
copy_df['loyer_median'] = copy_df['loyer_median'].astype(float, errors = 'raise')
copy_df['loyer_3_quartile'] = copy_df['loyer_3_quartile'].astype(float, errors = 'raise')
copy_df['loyer_9_decile'] = copy_df['loyer_9_decile'].astype(float, errors = 'raise')
copy_df['loyer_moyen'] = copy_df['loyer_moyen'].astype(float, errors = 'raise')
copy_df['loyer_mensuel_1_decile'] = copy_df['loyer_mensuel_1_decile'].astype(str).astype(float)
copy_df['loyer_mensuel_1_quartile'] = copy_df['loyer_mensuel_1_quartile'].astype(str).astype(float)
copy_df['loyer_mensuel_median'] = copy_df['loyer_mensuel_median'].astype(str).astype(float)
copy_df['loyer_mensuel_3_quartile'] = copy_df['loyer_mensuel_3_quartile'].astype(str).astype(float)
copy_df['loyer_mensuel_9_decile'] = copy_df['loyer_mensuel_9_decile'].astype(str).astype(float)
copy_df['moyenne_loyer_mensuel'] = copy_df['moyenne_loyer_mensuel'].astype(str).astype(float)
copy_df['surface_moyenne'] = copy_df['surface_moyenne'].astype(str).astype(float)
copy_df['nombre_observations'] = copy_df['nombre_observations'].astype(str).astype(float)
```

```python
dloyer1 = copy_df['loyer_1_decile'].mean()
qloyer1 = copy_df['loyer_1_quartile'].mean()
mloyer = copy_df['loyer_median'].mean()
qloyer3 = copy_df['loyer_3_quartile'].mean()
dloyer9 = copy_df['loyer_9_decile'].mean()
moloyer = copy_df['loyer_moyen'].mean()
lm1d = copy_df['loyer_mensuel_1_decile'].mean()
lm1q = copy_df['loyer_mensuel_1_quartile'].mean()
lmm = copy_df['loyer_mensuel_median'].mean()
lm3q = copy_df['loyer_mensuel_3_quartile'].mean()
lm9d = copy_df['loyer_mensuel_9_decile'].mean()
mlm = copy_df['moyenne_loyer_mensuel'].mean()
sm = copy_df['surface_moyenne'].mean()
no = copy_df['nombre_observations'].mean()
```

```python
copy_df['loyer_1_decile'].fillna(dloyer1, inplace=True)
copy_df['loyer_1_quartile'].fillna(qloyer1, inplace=True)
copy_df['loyer_median'].fillna(mloyer, inplace=True)
copy_df['loyer_3_quartile'].fillna(qloyer3, inplace=True)
copy_df['loyer_9_decile'].fillna(dloyer9, inplace=True)
copy_df['loyer_moyen'].fillna(moloyer, inplace=True)
copy_df['loyer_mensuel_1_decile'].fillna(lm1d, inplace=True)
copy_df['loyer_mensuel_1_quartile'].fillna(lm1q, inplace=True)
copy_df['loyer_mensuel_median'].fillna(lmm, inplace=True)
copy_df['loyer_mensuel_3_quartile'].fillna(lm3q, inplace=True)
copy_df['loyer_mensuel_9_decile'].fillna(lm9d, inplace=True)
copy_df['moyenne_loyer_mensuel'].fillna(mlm, inplace=True)
copy_df['surface_moyenne'].fillna(sm, inplace=True)
copy_df['nombre_observations'].fillna(no, inplace=True)
```

```python
copy_df = copy_df.round(2)
```
